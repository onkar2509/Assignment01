```python
#Import library 
import pandas as pd 
import seaborn as sns
#upload dataset 
df = pd.read_csv('/content/Students_Record.csv')
df.head
dat=pd.DataFrame(df)
dat
```

```python
df.shape
```

```python
#detection of null values 

df.isnull()
```

```python
#findout the null values

df.isnull().sum()
```

# Filling null value

```python
df2=df.fillna(value=0)
df2
```

```python
df2.isnull().sum().sum()
```

```python
#fill null value with previous value

df3=df.fillna(method='pad')
df3
```

```python
df3.isnull().sum().sum()
```

```python
#filling null value with next value

df4=df.fillna(method='bfill')
df4
```

```python
df4.isnull().sum().sum()
```

```python
#filling different value in Null different column

df5=df.fillna({'Maths_Score':'68',
               'Reading_Score':'89',
               'Writing_Score':'75',
               'Placement_Score':'65'})
df5
```

```python
df5.isnull().sum().sum()
```
