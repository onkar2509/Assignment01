# Scrape the tabular data from the following website: https://www.bankbazaar.com/goldrate/gold-rate-trend-in-india.html


```python
import bs4
from bs4 import BeautifulSoup as bs
import requests
import csv
```

```python
link="https://www.bankbazaar.com/gold-rate/gold-rate-trend-in-india.html"
```

```python
page = requests.get(link) 
```

```python
page.content
```

```python
soup = bs(page.content, 'html.parser')
print(soup.prettify())
```

# The scraped data should be stored in a CSV file.

```python
table = soup.find('table', {'class': 'table'})
```

```python
headers = [th.text.strip() for th in table.find_all('th')]
```

```python
rows = []
for tr in table.find_all('tr'):
    row = [td.text.strip() for td in tr.find_all('td')]
    if row:
        rows.append(row)
```

# The scraped data should be stored in a CSV file.


```python
with open('gold_data.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(rows)
```

# Train the Linear regression model using preprocessed data

```python
gold_data_updated = pd.read_csv('/content/gold_rates_updated.csv')
```

```python
X = gold_data_updated ['Year'].values.reshape(-1, 1)
y = gold_data_updated ['Price'].values.reshape(-1, 1)
```
